```python
import numpy as np
import os
from PIL import Image

from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

import joblib


# ==========================
# 1. Load MNIST
# ==========================

mnist = fetch_openml(
    "mnist_784",
    version=1
)

X = mnist.data.values
y = mnist.target.astype(int).values


# Normalize
X = X / 255.0


print("MNIST loaded:", X.shape)



# ==========================
# 2. Load your drawings
# ==========================

custom_images = []
custom_labels = []


base_path = "my_digits"


for digit in range(10):

    folder = os.path.join(
        base_path,
        str(digit)
    )

    for file in os.listdir(folder):

        if file.endswith(".png"):

            img = Image.open(
                os.path.join(folder,file)
            )

            img = img.convert("L")

            img = img.resize(
                (28,28)
            )

            arr = np.array(img)

            arr = arr.reshape(784)

            arr = arr / 255.0


            custom_images.append(arr)
            custom_labels.append(digit)



custom_images = np.array(custom_images)
custom_labels = np.array(custom_labels)


print(
    "Your images:",
    custom_images.shape
)



# ==========================
# 3. Combine datasets
# ==========================

X_final = np.concatenate(
    [
        X,
        custom_images
    ]
)


y_final = np.concatenate(
    [
        y,
        custom_labels
    ]
)


print(
    "Final dataset:",
    X_final.shape
)



# ==========================
# 4. Split data
# ==========================

X_train, X_test, y_train, y_test = train_test_split(
    X_final,
    y_final,
    test_size=0.2,
    random_state=42
)



# ==========================
# 5. Train model
# ==========================

model = RandomForestClassifier(
    n_estimators=150,
    random_state=42,
    n_jobs=-1
)


model.fit(
    X_train,
    y_train
)



# ==========================
# 6. Evaluate
# ==========================

prediction = model.predict(
    X_test
)


print(
    "Accuracy:",
    accuracy_score(
        y_test,
        prediction
    )
)


print(
    classification_report(
        y_test,
        prediction
    )
)



# ==========================
# 7. Save model
# ==========================

joblib.dump(
    model,
    "digit_classifier.pkl"
)


print("Model saved!")
```
